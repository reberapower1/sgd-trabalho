```python
#Start with imports
#To do: type your code here:
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

#Loading the iris data
#To do: type your code here:
data=load_iris()

print('Classes to predict: ', data.target_names)
#Result should be:
#Classes to predict:  ['setosa' 'versicolor' 'virginica']
```

### Exercise 2: Extracting data attributes (X) and target/ class labels (y)

```python
#To do: type your code here:

#To do: type your code here:
#Extracting data attributes
X =data.data

### Extracting target/ class labels
y = data.target

X[:4] #First 4 iris features
#results should be:
#array([[5.1, 3.5, 1.4, 0.2],
#       [4.9, 3. , 1.4, 0.2],
#       [4.7, 3.2, 1.3, 0.2],
#       [4.6, 3.1, 1.5, 0.2]])

print('Number of examples in the data:', y.shape[0]) ## result should be 150
```

### Exercise 3: Create train and test sets (use train_test_split), 25% for testing

```python
#To do: type your code here:
X_treino, X_teste= train_test_split(X, test_size=0.25, random_state=42)
y_treino, y_teste= train_test_split(y, test_size=0.25, random_state=42)
```

### Exercise 4: Create a Decision Tree (DT) classifier based on entropy (use DecisionTreeClassifier)

```python
#To do: type your code here:
clf = DecisionTreeClassifier(criterion='entropy')
```

### Exercise 5: Train the DT classifier

```python
#To do: type your code here:
clf.fit(X_treino, y_treino)
```

### Exercise 6: Predict labels on the test set.

```python
#To do: type your code here:

y_pred = clf.predict(X_teste)

print(y_pred.size) ## result should be 38
```

### Exercise 7: Evaluate results (accuracy on train and test set with accuracy_score)

```python
#To do: type your code here (fill the ...):

train_accuracy = accuracy_score(y_treino, clf.predict(X_treino))
test_accuracy = accuracy_score(y_teste, y_pred)

print('Train data accuracy: ', train_accuracy)
print('Test data accuracy: ', test_accuracy)

#Typical Output:
#Train data accuracy:  1.0
#Test data accuracy:  0.9736842105263158
```

### Exercise 8: Evaluate results (confusion matrix)

```python
#To do: type your code here (fill the ...):
cf_matrix = confusion_matrix(y_teste, y_pred)
print(cf_matrix)
```

### Exercise 9: Comment on results

```python
#This code generates a graphical confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')
ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Iris Category')
ax.set_ylabel('Actual Iris Category ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Setosa','Versicolor', 'Virginia'])
ax.yaxis.set_ticklabels(['Setosa','Versicolor', 'Virginia'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
